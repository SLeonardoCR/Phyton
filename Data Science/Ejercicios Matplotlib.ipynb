```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("datos_de_venta.csv")
print(df.head())
```

```python
#EJERCICIO 1
beneficio = str(df.columns[-1:][0])
plt.plot(list(df['numero_mes']), list(df['beneficio_total']), color='red', linestyle='dashed', linewidth=3, marker='o', markeredgecolor='black')
plt.legend(['Beneficio Total'], loc='lower right')
plt.plot
plt.xlabel('No. de Mes')
plt.ylabel('No. de Unidades Vendidas')
plt.show()
```

```python
#EJERCICIO 2
columns = list(df.columns)[0:-2]
#plt.plot(list(df['numero_mes']), list(df['beneficio_total']), color='red', linestyle='dashed', linewidth=3, marker='o', markeredgecolor='black')
colores = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', 'orange']
color = 0
for column in columns:
    #print(column)
    #print(list(df[column]))
    plt.plot(list(df['numero_mes']), list(df[column]), color=colores[color], linestyle='dashed', linewidth=3, marker='o', markeredgecolor='black')
    color = color + 1
plt.show()
```

```python
#EJERCICIO 3
bar1 = plt.bar(list(df['numero_mes']), list(df['crema_facial']), width = .9, color = 'purple', label = 'Crema Facial')
bar2 = plt.bar(list(df['numero_mes']), list(df['enjuague_facial']), width = .6, color = 'red', label = 'Enjueague Facial')
plt.legend(handles=[bar1, bar2], title = 'Ventas de:')
```

```python
#EJERCICIO 4
valores = []
columns = list(df.columns[1:-3])
for column in columns:
    valores.append(df[column].sum())
#print(valores)
#print(columns)
plt.pie(valores, labels = columns, autopct='%1.1f%%', startangle = 90)
```
