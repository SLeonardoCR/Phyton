```python
import requests
import re
from bs4 import BeautifulSoup
import concurrent.futures


# Realizar una solicitud GET a la página web
url = 'https://www.billboard.com/charts/year-end/2022/hot-100-songs/'
response = requests.get(url)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.5",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
}

# Comprobar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Analizar el contenido HTML de la página web
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Encontrar todos los elementos con la clase 'storylink' que contienen los títulos de los artículos
    titles = soup.find_all('h3', {'id':'title-of-a-story'})
    #print(titles)
    artists = soup.find_all('span', {'class':'lrv-u-font-size-14@mobile-max'})

    # Imprimir los títulos de los artículos
    
else:
    print('Error al realizar la solicitud.')
```

```python
def check_url(url, headers):
    try:
        response = requests.get(url, headers=headers, allow_redirects=True)
        # Si la página existe, el código de estado será 200
        #print(response.status_code)
        if response.status_code == 200:
            return True
        else:
            return False
    except requests.exceptions.RequestException as e:
        # Si hay un error al intentar acceder a la página, asumimos que no existe
        return False
        
def try_url(artist, title):
    url = 'https://www.musixmatch.com/lyrics/' + re.sub(r'\s+', '-', (artist  + '/' + title).replace('&', ''))
    if check_url(url, headers):
        print('checando ' + artist)
        return url
    else: 
        for i in range(1,11):
            print(' probando ' + artist + '-' + str(i))
            url = 'https://www.musixmatch.com/lyrics/' + re.sub(r'\s+', '-', (artist  + '-' + str(i) + '/' + title).replace('&', ''))
            if check_url(url, headers):
                return url
#print(check_url('https://www.musixmatch.com/lyrics/Glass-Animals/Heat-Waves', headers))
```

```python
#x = {}
#y = {}
all = []
for index, (title, artist) in enumerate(zip(titles, artists), start=1):
    if index >= 101:
        break
    #x[f"Title_{index}"] = title.text.replace('\n', '').replace('\t', '')
    #y[f"Artist_{index}"] = artist.text.replace('\n', '').replace('\t', '')
    all.append({'Position': index, 'Title': title.text.replace('\n', '').replace('\t', ''), 'Artist': artist.text.replace('\n', '').replace('\t', '')})
    
```

```python
def process_songs(all_songs):
    # Crea un ThreadPoolExecutor
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Crea una lista de futuros para cada llamada a try_url
        futures = [executor.submit(try_url, song['Artist'], song['Title']) for song in all_songs]
        # Espera a que todos los futuros se completen y recopila los resultados
        print('Cargando... jeje')
        results = [future.result() for future in concurrent.futures.as_completed(futures)]
    return results
```

```python
results = process_songs(all)
for result in results:
    print(result)
```
